```python
# import library soup & requests
from bs4 import BeautifulSoup
import requests
```

```python
# add url query params with example keyword
keyword = 'Pamungkas To The Bone'
url = 'https://www.youtube.com/results?search_query='+keyword.replace(' ',"+").lower()
```

```python
# tes print url param
url
```

```python
# response url 
response = requests.get(url)
```

```python
# parse html response to read
yt_soup = BeautifulSoup(data,'html.parser')
```
